```python
#Add a "Grade" column (e.g., A for >=270, B for 240269, etc.)
import pandas as np
df=pd.read_csv('myproject1.csv')
df['Total']=df[['Math','Science','English']].sum(axis=1)
def assign_grade(total):
  if total>=270:
    return 'A'
  elif total>=240:
    return 'B'
  elif total>=210:
    return 'C'
  else:
    return 'D'
df['Grade']=df['Total'].apply(assign_grade)
print(df[['Name','Total','Grade']])
```

```python
#sort the students accord to grade
import pandas as np
df=pd.read_csv('myproject1.csv')
df['Total']=df[['Math','Science','English']].sum(axis=1)
def assign_grade(total):
  if total>=270:
    return 'A'
  elif total>=240:
    return 'B'
  elif total>=210:
    return 'C'
  else:
    return 'D'
df['Grade']=df['Total'].apply(assign_grade)
p=df.sort_values(by='Grade',ascending=True)
print(p[['Name','Grade']])
```

```python
#Find the top 3 students in Math, Science, and English separately.
top_math = df.sort_values(by='Math', ascending=False).head(3)
print(" Top 3 Students in Math:")
print(top_math[['Name', 'Math']])
print()
top_Science = df.sort_values(by='Science', ascending=False).head(3)
print(" Top 3 Students in Science:")
print(top_Science[['Name', 'Science']])
print()
top_English= df.sort_values(by='English', ascending=False).head(3)
print(" Top 3 Students in English:")
print(top_math[['Name', 'English']])
print()
```

```python
#Create a column showing if a student passed all subjects (e.g., 40+ marks).
import pandas as pd
df=pd.read_csv('myproject1.csv')
data=pd.DataFrame(df)
data['Passed_all']=df[['Math','Science','English']].ge(40).all(axis=1)
print(data)
```

```python
#Who has the most consistent performance across all subjects? (Smallest difference between highest and lowest subject marks)
import pandas as pd
data=pd.read_csv('myproject1.csv')
df = pd.DataFrame(data)

# Calculate the range (max - min) for each student
df['Score_Range'] = df[['Math', 'Science', 'English']].max(axis=1) - df[['Math', 'Science', 'English']].min(axis=1)

# Find the most consistent performer (minimum range)
most_consistent = df.loc[df['Score_Range'].idxmin()]

print("Most Consistent Performer:")
print(most_consistent[['Name', 'Score_Range']])
```
